```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, cross_validate
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier, VotingClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')


RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)


def load_data(path='diabetes (1).csv'):
    df = pd.read_csv(path)
    return df


def preprocess(df):
    df = df.copy()
    # Columns where zero is invalid
    zero_cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
    for c in zero_cols:
        df[c] = df[c].replace(0, np.nan)
    imputer = SimpleImputer(strategy='median')
    df[zero_cols] = imputer.fit_transform(df[zero_cols])
    X = df.drop('Outcome', axis=1)
    y = df['Outcome']
    return X, y


def evaluate_model(clf, X_train, X_test, y_train, y_test):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return acc, prec, rec, f1


def cross_val_metrics(clf, X, y, cv=10):
    scoring = ['accuracy','precision','recall','f1']
    scores = cross_validate(clf, X, y, cv=cv, scoring=scoring, n_jobs=-1)
    return {k: np.mean(scores['test_'+k]) for k in scoring}


def main():
    if not os.path.exists('diabetes (1).csv'):
        raise FileNotFoundError('Place diabetes (1).csv in the working directory')
    df = load_data('diabetes (1).csv')
    print('Dataset shape:', df.shape)
    X, y = preprocess(df)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE)


    # Scaling for SVM
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)


    results = []


    # Decision Tree
    dt = DecisionTreeClassifier(criterion='gini', random_state=RANDOM_STATE)
    dt_acc, dt_prec, dt_rec, dt_f1 = evaluate_model(dt, X_train, X_test, y_train, y_test)
    dt_cv = cross_val_metrics(dt, X, y)
    results.append(('Decision Tree', dt_acc, dt_prec, dt_rec, dt_f1, dt_cv))


    # SVM (requires scaling)
    svm = SVC(kernel='rbf', C=1.0, gamma='scale', probability=True, random_state=RANDOM_STATE)
    svm_acc, svm_prec, svm_rec, svm_f1 = evaluate_model(svm, X_train_scaled, X_test_scaled, y_train, y_test)
    svm_cv = cross_val_metrics(svm, scaler.fit_transform(X), y)
    results.append(('SVM', svm_acc, svm_prec, svm_rec, svm_f1, svm_cv))


    # Random Forest
    rf = RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE)
    rf_acc, rf_prec, rf_rec, rf_f1 = evaluate_model(rf, X_train, X_test, y_train, y_test)
    rf_cv = cross_val_metrics(rf, X, y)
    results.append(('Random Forest', rf_acc, rf_prec, rf_rec, rf_f1, rf_cv))


    # Proposed: RFE + HistGradientBoosting
    selector_est = RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE)
    rfe = RFE(estimator=selector_est, n_features_to_select=5)
    rfe.fit(X_train, y_train)
    selected_cols = X_train.columns[rfe.support_].tolist()
    print('Selected features by RFE:', selected_cols)
    X_train_sel = rfe.transform(X_train)
    X_test_sel = rfe.transform(X_test)


    hgb = HistGradientBoostingClassifier(max_iter=200, learning_rate=0.1, max_leaf_nodes=31, random_state=RANDOM_STATE)
    hgb_acc, hgb_prec, hgb_rec, hgb_f1 = evaluate_model(hgb, X_train_sel, X_test_sel, y_train, y_test)
    hgb_cv = cross_val_metrics(hgb, rfe.transform(X), y)
    results.append(('RFE+HGB', hgb_acc, hgb_prec, hgb_rec, hgb_f1, hgb_cv))


       # --- Save results ---
    rows = []
    for model, acc, prec, rec, f1, cv in results:
        rows.append({
            'Model': model,
            'Accuracy': round(acc,3),
            'Precision': round(prec,3),
            'Recall': round(rec,3),
            'F1': round(f1,3),
            'CV_Accuracy': round(cv['accuracy'],3),
            'CV_Precision': round(cv['precision'],3),
            'CV_Recall': round(cv['recall'],3),
            'CV_F1': round(cv['f1'],3)
        })

    results_df = pd.DataFrame(rows)
    print("\n Final Results:")
    print(results_df.to_string(index=False))
    results_df.to_csv('model_results.csv', index=False)
    print("\n Results saved to model_results.csv")

    # --- Bar chart ---
    plt.figure(figsize=(8,5))
    plt.bar(results_df['Model'], results_df['Accuracy'], color='skyblue', label='Accuracy')
    plt.bar(results_df['Model'], results_df['CV_Accuracy'], color='orange', alpha=0.5, label='CV Accuracy')
    plt.ylabel('Accuracy')
    plt.title('Model Comparison')
    plt.legend()
    plt.tight_layout()
    plt.savefig('model_comparison.png')
    plt.show()
    print("Bar chart saved as model_comparison.png")


if __name__ == "__main__":
    main()
```
